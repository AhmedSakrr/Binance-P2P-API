```python
from BP2P import get_all_ads
import pandas as pd 
from datetime import datetime
```

```python
get_all_ads('USDT', 'SELL')
```

```python
get_all_ads('USDT', 'BUY')
```

```python
sell_usdt = get_all_ads('USDT', 'SELL')
buy_usdt = get_all_ads('USDT', 'BUY')
arbitrageurs = pd.merge(buy_usdt, sell_usdt, on=['advertiser.nickName'], how='inner')
data = arbitrageurs[['advertiser.nickName', 'adv.price_x', 'adv.price_y', 'advertiser.monthOrderCount_x']]
data
```

```python
now = datetime.now()
current_time = now.strftime("%d/%m/%Y, %H:%M:%S")

ax = sell_usdt['adv.price'].astype(float).plot()
buy_usdt['adv.price'].astype(float).plot(ax=ax, figsize=(15,6), title='Adv Price')
ax.set_ylabel("Price in MAD")   
ax.set_xlabel("Ad index")    

arbitrageurs['adv.price_x'].astype(float).plot(ax=ax)
arbitrageurs['adv.price_y'].astype(float).plot(ax=ax, figsize=(20,10), title='USDT Market '+current_time)
ax.set_ylabel("Price in MAD")   
ax.set_xlabel("Ad Index")   
ax.legend(["If you wanna Sell in market ", "if you wanna Buy in market","Where Arbitrageurs wanna Sell", "Where Arbitrageurs wanna Buy"]);     
```

```python
ax = sell_usdt['adv.minSingleTransAmount'].astype(float).plot()
buy_usdt['adv.minSingleTransAmount'].astype(float).plot(ax=ax, figsize=(15,6),title='Min Single Transaction Amount ')
ax.set_ylabel("Amount in MAD")    
ax.set_xlabel("Ad index")  
ax.legend(["If you wanna Sell", "if you wanna Buy"]);   
```

```python
ax = arbitrageurs['adv.minSingleTransAmount_x'].astype(float).plot()
arbitrageurs['adv.minSingleTransAmount_y'].astype(float).plot(ax=ax, figsize=(15,6),title='Min Single Transaction Amount for Arbitrageurs')
ax.set_ylabel("Amount in MAD")    
ax.set_xlabel("Ad index")  
ax.legend(["If you wanna Sell to them ", "if you wanna Buy from them"]);   
```
